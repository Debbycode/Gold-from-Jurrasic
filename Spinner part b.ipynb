```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataframe = pd.read_excel("New Microsoft Office Excel Worksheet.xlsx")
```

```python
A = dataframe[["Vcab-129","Vcab-96","Vcab-62","Vcab-33","Vcab-33"]]
B = dataframe[["Vcab32", "Vcab61", "Vcab93" ,"Vcab125"]]
plt.figure(figsize=(16,16))

ax1 = plt.subplot(1,2,1)
plt.plot( A,dataframe["Depth"])
plt.ylim(11200,10300)
plt.xlabel("Spinner Response(rev./second)")
plt.ylabel("Depth(feet)")
plt.title("Logged Up")
plt.legend(A,loc = "best")
plt.grid()


ax2 = plt.subplot(1,2,2)
plt.plot( B,dataframe["Depth"])
plt.ylim(11200,10300)
plt.xlabel("Spinner Response(rev./second)")
plt.ylabel("Depth(feet)")
plt.title("Logged Down")
plt.legend(B,loc = "best")
plt.grid()
```

```python
CableSpeed = [-129,-96,-62,-33,32,61,93,125]
```

```python
dataframe["Slope"] = (dataframe["Vcab32"]-dataframe["Vcab-33"])/65
```

```python
dataframe["f0"] = (dataframe["Vcab32"]+dataframe["Vcab-33"])/2
```

```python
VTHRESH = 8.786862003780746
```

```python
dataframe["vf"] = (dataframe["f0"]/dataframe["Slope"]) + VTHRESH
```

```python
dataframe.set_index("Station")
```

```python
dataframe["flow%"] = dataframe["vf"]/dataframe["vf"][0] * 100
```

```python
plt.figure(figsize=(4,8))
plt.plot(dataframe["flow%"],dataframe["Depth"],c = "red")
plt.ylim(11200,10300)
plt.xlabel("Percent Total Flow")
plt.ylabel("Depth(ft)")
plt.title("Multipass Spinner Method Flow Profile")
plt.grid()
```
