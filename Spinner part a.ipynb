```python
dataframe["Slope"] = (dataframe["Vcab32"]-dataframe["Vcab-33"])/65
```

```python
dataframe
```

```python
dataframe["f0"] = (dataframe["Vcab32"]+dataframe["Vcab-33"])/2
```

```python
dataframe
```

```python
#for threshhold velocity taking values from station 13 at depth 11020 feet
fr_neg = [-6,-3.5,-1.5,-1]
fr_pos = [4,5.5,7.5,9]
v_neg = [-129,-96,-62,-33]
v_pos = [32,61,93,125]
plt.plot(v_neg,fr_neg)
plt.plot(v_pos,fr_pos)
plt.grid()
```

```python
m,b = np.polyfit(v_neg,fr_neg,1)
m1,b1 = np.polyfit(v_pos,fr_pos,1)
Vc0_pos = -b/m
Vc0_neg = -b1/m1
Vthreshold = abs(Vc0_neg - Vc0_pos)/2
print("The threshold velocity is", Vthreshold )
```

```python
dataframe["vf"] = (dataframe["f0"]/dataframe["Slope"]) + Vthreshold
```

```python
dataframe
```

```python
dataframe["flow%"] = dataframe["vf"]/dataframe["vf"]["station1"] * 100
```

```python
dataframe
```

```python
plt.figure(figsize=(4,8))
plt.plot(dataframe["flow%"],dataframe["Depth"],c = "red")
plt.ylim(11200,10300)
plt.xlabel("Percent Total Flow")
plt.ylabel("Depth(ft)")
plt.title("Multipass Spinner Method Flow Profile")
plt.grid()
```
